```python
import pandas as pd
import numpy as np
import timeit
```

# The Data Processing

```python
# 数据集处理函数
def data_process(data_set, feature_size):
    total=[]
    dict_1={}
    for i in range(feature_size):
        a=str(i+1)
        dict_1[a]=0.0
    total.append(dict_1)
    for i in range(len(data_set)):
        data_sep=[]
        data_sep.append(data_set[i][0].split(' '))
        data_sep[0] = list(filter(None, data_sep[0]))
        new_dict={}
        new_dict['label']=float(data_sep[0][0])
        for j in range(len(data_sep[0])-1):
            a=[]
            a.append(data_sep[0][j+1].split(':'))
            new_dict[a[0][0]]=float(a[0][1])
        total.append(new_dict)
    df=pd.DataFrame(total).fillna(0.0)
    df=df.drop(index=(df.loc[(df['label']==0.0)].index))
    return df
```

```python
# 处理训练集
ijcnn1_tr ='ijcnn1.tr'
ijcnn1_tr_file = open(ijcnn1_tr,'r')
lines=ijcnn1_tr_file.readlines()

ijcnn1_tr_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    ijcnn1_tr_data.append(temp2)


ijcnn1_tr_data_process = data_process(ijcnn1_tr_data, 22)
ijcnn1_tr_data_process
```

```python
type(ijcnn1_tr_data_process['label'][1])
```

```python
# 处理测试集
ijcnn1_t ='ijcnn1.t'
ijcnn1_t_file = open(ijcnn1_t,'r')
lines=ijcnn1_t_file.readlines()

ijcnn1_t_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    ijcnn1_t_data.append(temp2)
ijcnn1_t_data

ijcnn1_t_data_process = data_process(ijcnn1_t_data, 22)
ijcnn1_t_data_process
```

```python
# In the search mission, we dont need the label, So we get a dataset without label
ijcnn1_t_data_process_No_label = ijcnn1_t_data_process.drop('label',axis = 1)
ijcnn1_t_data_process_No_label
```

# Test for Search speed

### First test for ball-tree

```python
# Now we ues the KNN balltree to test the time for Search
# We use the default parameters
import numpy as np
from sklearn.neighbors import BallTree

# 1.Euclidean  2.leaf_sizepositive:default=40
start=timeit.default_timer()
tree = BallTree(ijcnn1_t_data_process_No_label, leaf_size=2)   
stop=timeit.default_timer()
build_tree_time = stop-start
print("The time cost on build the tree")
print(build_tree_time)

print("Now we start to test a search in this ball-tree")
print("We select the index=2 datapoint , to find its 3 neighbors ")
start=timeit.default_timer()
dist, ind = tree.query(ijcnn1_t_data_process_No_label.iloc[[2]], k=4)
stop=timeit.default_timer()
print("We find the 3 cloest neighbors (the result include the data themself), the index are below ")
print(ind)  # indices of 3 closest neighbors
print("We find the 3 cloest neighbors , the distance are below ")
print(dist)  # distances to 3 closest neighbors
print("The search time cost is that ")
run_time = stop - start
print(run_time)

```

```python
tree = BallTree(ijcnn1_t_data_process_No_label, leaf_size=2) 
every_time = []
for i in range(10000):
    start=timeit.default_timer()
    dist, ind = tree.query(ijcnn1_t_data_process_No_label.iloc[[i]], k=4)
    stop=timeit.default_timer()
    time_cost = stop- start
    every_time.append(time_cost)
```

```python
run_search_time_balltree = np.mean(every_time)
run_search_time_balltree
```

### Second test for Kd-tree

```python
# Now we ues the KNN kdtree to test the time for Search
# We use the default parameters
import numpy as np
from sklearn.neighbors import KDTree

# 1.Euclidean  2.leaf_sizepositive:default=40
start=timeit.default_timer()
tree = KDTree(ijcnn1_t_data_process_No_label, leaf_size=2)   
stop=timeit.default_timer()
build_tree_time = stop-start
print("The time cost on build the tree")
print(build_tree_time)

print("Now we start to test a search in this kd-tree")
print("We select the index=2 datapoint , to find its 3 neighbors ")
start=timeit.default_timer()
dist, ind = tree.query(ijcnn1_t_data_process_No_label.iloc[[2]], k=4)
stop=timeit.default_timer()
print("We find the 3 cloest neighbors (the result include the data themself), the index are below ")
print(ind)  # indices of 3 closest neighbors
print("We find the 3 cloest neighbors , the distance are below ")
print(dist)  # distances to 3 closest neighbors
print("The search time cost is that ")
run_time = stop - start
print(run_time)
print("The neighbors of index=2 datapoint are index=[472,62,36319]")
```

```python
tree = KDTree(ijcnn1_t_data_process_No_label, leaf_size=2) 
every_time = []
for i in range(10000):
    start=timeit.default_timer()
    dist, ind = tree.query(ijcnn1_t_data_process_No_label.iloc[[i]], k=4)
    stop=timeit.default_timer()
    time_cost = stop- start
    every_time.append(time_cost)
```

```python
run_search_time_KDtree = np.mean(every_time)
run_search_time_KDtree
```

The search speed of kdtree is fast than balltree
