```python
import pandas as pd
import numpy as np
import timeit
```

```python
# 数据集处理函数
def data_process(data_set, feature_size):
    total=[]
    dict_1={}
    for i in range(feature_size):
        a=str(i+1)
        dict_1[a]=0.0
    total.append(dict_1)
    for i in range(len(data_set)):
        data_sep=[]
        data_sep.append(data_set[i][0].split(' '))
        data_sep[0] = list(filter(None, data_sep[0]))
        new_dict={}
        new_dict['label']=float(data_sep[0][0])
        for j in range(len(data_sep[0])-1):
            a=[]
            a.append(data_sep[0][j+1].split(':'))
            new_dict[a[0][0]]=float(a[0][1])
        total.append(new_dict)
    df=pd.DataFrame(total).fillna(0.0)
    df=df.drop(index=(df.loc[(df['label']==0.0)].index))
    return df
```

```python
# 处理训练集
mnist_tr ='mnist.scale'
mnist_tr_file = open(mnist_tr,'r')
lines = mnist_tr_file.readlines()

mnist_tr_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    mnist_tr_data.append(temp2)


mnist_tr_data_process = data_process(mnist_tr_data, 784)
mnist_tr_data_process
```

```python
# 处理测试集
mnist_t ='mnist.scale.t'
mnist_t_file = open(mnist_t,'r')
lines = mnist_t_file.readlines()

mnist_t_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    mnist_t_data.append(temp2)


mnist_t_data_process = data_process(mnist_t_data, 784)
mnist_t_data_process
```

```python
mnist_t_data_process_nolabel = mnist_t_data_process.drop('label',axis=1)
```

# Test for Search speed

### First test for ball-tree

```python
# Now we ues the KNN balltree to test the time for Search
# We use the default parameters
import numpy as np
from sklearn.neighbors import BallTree

# 1.Euclidean  2.leaf_sizepositive:default=40
start=timeit.default_timer()
tree = BallTree(mnist_t_data_process_nolabel, leaf_size=2)   
stop=timeit.default_timer()
build_tree_time = stop-start
print("The time cost on build the tree")
print(build_tree_time)

print("Now we start to test a search in this ball-tree")
print("We select the index=2 datapoint , to find its 3 neighbors ")
start=timeit.default_timer()
dist, ind = tree.query(mnist_t_data_process_nolabel.iloc[[2]], k=4)
stop=timeit.default_timer()
print("We find the 3 cloest neighbors (the result include the data themself), the index are below ")
print(ind)  # indices of 3 closest neighbors
print("We find the 3 cloest neighbors , the distance are below ")
print(dist)  # distances to 3 closest neighbors
print("The search time cost is that ")
run_time = stop - start
print(run_time)

```

```python
tree = BallTree(mnist_t_data_process_nolabel, leaf_size=2) 
every_time = []
for i in range(5000):
    start=timeit.default_timer()
    dist, ind = tree.query(mnist_t_data_process_nolabel.iloc[[i]], k=4)
    stop=timeit.default_timer()
    time_cost = stop- start
    every_time.append(time_cost)
```

```python
Search_mean_time_balltree = np.mean(every_time)
Search_mean_time_balltree
```

### Second test for KD-tree

```python
# Now we ues the KNN kdtree to test the time for Search
# We use the default parameters
import numpy as np
from sklearn.neighbors import KDTree

# 1.Euclidean  2.leaf_sizepositive:default=40
start=timeit.default_timer()
tree = KDTree(mnist_t_data_process_nolabel, leaf_size=2)   
stop=timeit.default_timer()
build_tree_time = stop-start
print("The time cost on build the tree")
print(build_tree_time)

print("Now we start to test a search in this kd-tree")
print("We select the index=2 datapoint , to find its 3 neighbors ")
start=timeit.default_timer()
dist, ind = tree.query(mnist_t_data_process_nolabel.iloc[[2]], k=4)
stop=timeit.default_timer()
print("We find the 3 cloest neighbors (the result include the data themself), the index are below ")
print(ind)  # indices of 3 closest neighbors
print("We find the 3 cloest neighbors , the distance are below ")
print(dist)  # distances to 3 closest neighbors
print("The search time cost is that ")
run_time = stop - start
print(run_time)

```

```python
tree = KDTree(mnist_t_data_process_nolabel, leaf_size=2) 
every_time = []
for i in range(5000):
    start=timeit.default_timer()
    dist, ind = tree.query(mnist_t_data_process_nolabel.iloc[[i]], k=4)
    stop=timeit.default_timer()
    time_cost = stop- start
    every_time.append(time_cost)
```

```python
Search_mean_time_kdtree = np.mean(every_time)
Search_mean_time_kdtree
```

The search speed of balltree is significant fast than kdtree!

# Test for classify

```python
# Now we test the Classifiy on the KNN BallTree
```

```python
X = mnist_tr_data_process.drop('label',axis=1)
X
```

```python
Y = mnist_tr_data_process['label']
Y
```

```python
test_X = mnist_t_data_process.drop('label',axis=1)
test_X
```

```python
# Now we test the Classifiy on the KNN balltree
# Parameter:
#  n_neighbors=3
#  algorithm:balltree
#  leaf_sizeint, default=30
#  pint, default=2 ,it means we used the euclidean_distance
from sklearn.neighbors import KNeighborsClassifier
start=timeit.default_timer()
neigh = KNeighborsClassifier(n_neighbors=3,algorithm ='ball_tree')
neigh.fit(X, Y)
predict_result = neigh.predict(test_X)
stop=timeit.default_timer()
run_classfy_time_balltree = stop - start
print("run_classfy_time_balltree")
print(run_classfy_time_balltree)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_true = mnist_t_data_process['label'].values
y_pred = predict_result
accuracy_score(y_true, y_pred)
# The accuracy 
```

```python
print(classification_report(y_true, y_pred))
```

```python
# Now we test the Classifiy on the KNN kdTree
# Parameter:
#  n_neighbors=3
#  algorithm:kdtree
#  leaf_sizeint, default=30
#  pint, default=2 ,it means we used the euclidean_distance
from sklearn.neighbors import KNeighborsClassifier
start=timeit.default_timer()
neigh = KNeighborsClassifier(n_neighbors=3,algorithm ='kd_tree')
neigh.fit(X, Y)
predict_result = neigh.predict(test_X)
stop=timeit.default_timer()
run_classfy_time_kdtree = stop - start
print("run_classfy_time_kdtree")
print(run_classfy_time_kdtree)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_true = mnist_t_data_process['label'].values
y_pred = predict_result
accuracy_score(y_true, y_pred)
# The accuracy 
```

```python
accuracy = accuracy_score(y_true, y_pred)
```

```python
print(classification_report(y_true, y_pred))
```

```python
result = pd.DataFrame(columns=['Classify time','Classify Accuracy','Search means time'])
result.loc['balltree'] = [run_classfy_time_balltree,accuracy,Search_mean_time_balltree]
result.loc['kdtree'] = [run_classfy_time_kdtree,accuracy,Search_mean_time_kdtree]
```
