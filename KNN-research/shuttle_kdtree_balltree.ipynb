```python
import pandas as pd
import numpy as np
import timeit
```

```python
# Data process
def data_process(data_set, feature_size):
    total=[]
    dict_1={}
    for i in range(feature_size):
        a=str(i+1)
        dict_1[a]=0.0
    total.append(dict_1)
    for i in range(len(data_set)):
        data_sep=[]
        data_sep.append(data_set[i][0].split(' '))
        data_sep[0] = list(filter(None, data_sep[0]))
        new_dict={}
        new_dict['label']=float(data_sep[0][0])
        for j in range(len(data_sep[0])-1):
            a=[]
            a.append(data_sep[0][j+1].split(':'))
            new_dict[a[0][0]]=float(a[0][1])
        total.append(new_dict)
    df=pd.DataFrame(total).fillna(0.0)
    df=df.drop(index=(df.loc[(df['label']==0.0)].index))
    return df
```

```python
# Train data
shuttle_tr ='shuttle.scale.tr'
shuttle_tr_file = open(shuttle_tr,'r')
lines = shuttle_tr_file.readlines()

shuttle_tr_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    shuttle_tr_data.append(temp2)


shuttle_tr_data_process = data_process(shuttle_tr_data, 9)
shuttle_tr_data_process
```

```python
# test data
shuttle_t ='shuttle.scale.t'
shuttle_t_file = open(shuttle_t,'r')
lines = shuttle_t_file.readlines()

shuttle_t_data=[]
for line in lines:
    temp1=line.strip('\n')
    temp2=temp1.split('\t')
    shuttle_t_data.append(temp2)


shuttle_t_data_process = data_process(shuttle_t_data, 9)
shuttle_t_data_process
```

```python
shuttle_t_data_process_nolabel = shuttle_t_data_process.drop('label',axis=1)
```

# Test for classify

```python
X = shuttle_tr_data_process.drop('label',axis=1)
X
```

```python
Y = shuttle_tr_data_process['label']
Y
```

```python
test_X = shuttle_t_data_process.drop('label',axis=1)
test_X
```

```python
# Now we test the Classifiy on the KNN balltree
# Parameter:
#  n_neighbors = 3
#  algorithm:balltree
#  leaf_sizeint, default=30
#  pint, default=2 ,it means we used the euclidean_distance
from sklearn.neighbors import KNeighborsClassifier
start=timeit.default_timer()
neigh = KNeighborsClassifier(n_neighbors=3,algorithm ='ball_tree')
neigh.fit(X, Y)
predict_result = neigh.predict(test_X)
stop=timeit.default_timer()
run_classfy_time_balltree = stop - start
print("run_classfy_time_balltree")
print(run_classfy_time_balltree)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_true = shuttle_t_data_process['label'].values
y_pred = predict_result
accuracy_score(y_true, y_pred)
# The accuracy 
```

```python
print(classification_report(y_true,y_pred))
```

```python
# Now we test the Classifiy on the KNN kdTree
# Parameter:
#  n_neighbors = 3
#  algorithm:kd-tree
#  leaf_sizeint, default=30
#  pint, default=2 ,it means we used the euclidean_distance
from sklearn.neighbors import KNeighborsClassifier
start=timeit.default_timer()
neigh = KNeighborsClassifier(n_neighbors=3,algorithm ='kd_tree')
neigh.fit(X, Y)
predict_result = neigh.predict(test_X)
stop=timeit.default_timer()
run_classfy_time_kdtree = stop - start
print("run_classfy_time_kdtree")
print(run_classfy_time_kdtree)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_true = shuttle_t_data_process['label'].values
y_pred = predict_result
accuracy_score(y_true, y_pred)
# The accuracy 
```

```python
print(classification_report(y_true,y_pred))
```

```python
accuracy = accuracy_score(y_true, y_pred) 
```
